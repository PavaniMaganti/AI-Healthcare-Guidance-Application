```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix
```

```python
df = pd.read_csv("healthcare_training_dataset.csv")

df.head()
```

```python
print("Dataset shape:", df.shape)
print("Number of columns:", len(df.columns))
print("First 10 columns:")
print(df.columns[:10])
print("Last 5 columns:")
print(df.columns[-5:])
```

```python
print("Number of diseases:", df["disease"].nunique())
print(df["disease"].value_counts().head(10))
```

```python
print("Dataset shape:", df.shape)

print("\nTotal missing values:")
print(df.isnull().sum().sum())

print("\nDuplicate rows:")
print(df.duplicated().sum())

print("\nDisease count minimum:", df["disease"].value_counts().min())
print("Disease count maximum:", df["disease"].value_counts().max())

symptom_columns = [col for col in df.columns if col != "disease"]

print("\nNumber of symptom columns:", len(symptom_columns))

# Check whether symptom columns only contain 0 and 1
valid_binary = df[symptom_columns].isin([0, 1]).all().all()
print("Are all symptom columns binary 0/1?", valid_binary)
```

```python
df["disease"] = df["disease"].replace({
    "Peptic ulcer diseae": "Peptic ulcer disease"
})

print(df["disease"].unique())
```

```python
print("Before removing duplicates:", df.shape)

df = df.drop_duplicates()

print("After removing duplicates:", df.shape)
print("Duplicates remaining:", df.duplicated().sum())
```

```python
print("Number of diseases:", df["disease"].nunique())

disease_counts = df["disease"].value_counts()

print("Minimum examples per disease:", disease_counts.min())
print("Maximum examples per disease:", disease_counts.max())

disease_counts
```

```python
# Features = symptom columns
X = df.drop(columns=["disease"])

# Target = disease column
y = df["disease"]

print("X shape:", X.shape)
print("y shape:", y.shape)

print("\nFirst 5 target values:")
print(y.head())
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

y_encoded = label_encoder.fit_transform(y)

print("Original disease names:")
print(y.head())

print("\nEncoded disease labels:")
print(y_encoded[:10])

print("\nNumber of classes:", len(label_encoder.classes_))

print("\nFirst 10 disease classes:")
print(label_encoder.classes_[:10])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y_encoded,
    test_size=0.2,
    random_state=42,
    stratify=y_encoded
)

print("Training data:", X_train.shape)
print("Testing data:", X_test.shape)
print("Training labels:", y_train.shape)
print("Testing labels:", y_test.shape)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

# Create SVM model
svm_model = SVC(kernel="linear", probability=True, random_state=42)

# Train the model
svm_model.fit(X_train, y_train)

# Predict on test data
svm_predictions = svm_model.predict(X_test)

# Evaluate
svm_accuracy = accuracy_score(y_test, svm_predictions)
svm_precision = precision_score(y_test, svm_predictions, average="weighted", zero_division=0)
svm_recall = recall_score(y_test, svm_predictions, average="weighted", zero_division=0)
svm_f1 = f1_score(y_test, svm_predictions, average="weighted", zero_division=0)

print("SVM Model Results")
print("Accuracy:", round(svm_accuracy, 4))
print("Precision:", round(svm_precision, 4))
print("Recall:", round(svm_recall, 4))
print("F1 Score:", round(svm_f1, 4))
```

```python
from sklearn.naive_bayes import MultinomialNB

# Create Naive Bayes model
nb_model = MultinomialNB()

# Train the model
nb_model.fit(X_train, y_train)

# Predict on test data
nb_predictions = nb_model.predict(X_test)

# Evaluate
nb_accuracy = accuracy_score(y_test, nb_predictions)
nb_precision = precision_score(y_test, nb_predictions, average="weighted", zero_division=0)
nb_recall = recall_score(y_test, nb_predictions, average="weighted", zero_division=0)
nb_f1 = f1_score(y_test, nb_predictions, average="weighted", zero_division=0)

print("Naive Bayes Model Results")
print("Accuracy:", round(nb_accuracy, 4))
print("Precision:", round(nb_precision, 4))
print("Recall:", round(nb_recall, 4))
print("F1 Score:", round(nb_f1, 4))
```

```python
from sklearn.ensemble import RandomForestClassifier

# Create Random Forest model
rf_model = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    class_weight="balanced"
)

# Train the model
rf_model.fit(X_train, y_train)

# Predict on test data
rf_predictions = rf_model.predict(X_test)

# Evaluate
rf_accuracy = accuracy_score(y_test, rf_predictions)
rf_precision = precision_score(y_test, rf_predictions, average="weighted", zero_division=0)
rf_recall = recall_score(y_test, rf_predictions, average="weighted", zero_division=0)
rf_f1 = f1_score(y_test, rf_predictions, average="weighted", zero_division=0)

print("Random Forest Model Results")
print("Accuracy:", round(rf_accuracy, 4))
print("Precision:", round(rf_precision, 4))
print("Recall:", round(rf_recall, 4))
print("F1 Score:", round(rf_f1, 4))
```

```python
model_results = pd.DataFrame({
    "Model": ["SVM", "Naive Bayes", "Random Forest"],
    "Accuracy": [svm_accuracy, nb_accuracy, rf_accuracy],
    "Precision": [svm_precision, nb_precision, rf_precision],
    "Recall": [svm_recall, nb_recall, rf_recall],
    "F1 Score": [svm_f1, nb_f1, rf_f1]
})

model_results = model_results.sort_values(by="F1 Score", ascending=False)

model_results
```

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

models = {
    "SVM": svm_model,
    "Naive Bayes": nb_model,
    "Random Forest": rf_model
}

cv_results = []

for model_name, model in models.items():
    scores = cross_val_score(
        model,
        X,
        y_encoded,
        cv=cv,
        scoring="f1_weighted"
    )

    cv_results.append({
        "Model": model_name,
        "CV F1 Mean": scores.mean(),
        "CV F1 Std": scores.std()
    })

cv_results_df = pd.DataFrame(cv_results).sort_values(
    by="CV F1 Mean",
    ascending=False
)

cv_results_df
```

```python
print("Random Forest Classification Report")
print(classification_report(
    y_test,
    rf_predictions,
    target_names=label_encoder.classes_,
    zero_division=0
))
```

```python
# Clean disease names for professional app output
df["disease"] = df["disease"].str.strip()

df["disease"] = df["disease"].replace({
    "Peptic ulcer diseae": "Peptic ulcer disease",
    "(vertigo) Paroymsal  Positional Vertigo": "Paroxysmal Positional Vertigo",
    "Dimorphic hemmorhoids(piles)": "Haemorrhoids / Piles",
    "Osteoarthristis": "Osteoarthritis",
    "hepatitis A": "Hepatitis A"
})

print("Number of diseases:", df["disease"].nunique())
print(df["disease"].sort_values().unique())
```

```python
X = df.drop(columns=["disease"])
y = df["disease"]

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y_encoded,
    test_size=0.2,
    random_state=42,
    stratify=y_encoded
)

print("X shape:", X.shape)
print("y shape:", y.shape)
print("Training data:", X_train.shape)
print("Testing data:", X_test.shape)
print("Number of disease classes:", len(label_encoder.classes_))
```

```python
final_model = MultinomialNB()

final_model.fit(X_train, y_train)

final_predictions = final_model.predict(X_test)

final_accuracy = accuracy_score(y_test, final_predictions)
final_precision = precision_score(y_test, final_predictions, average="weighted", zero_division=0)
final_recall = recall_score(y_test, final_predictions, average="weighted", zero_division=0)
final_f1 = f1_score(y_test, final_predictions, average="weighted", zero_division=0)

print("Final Naive Bayes Model Results")
print("Accuracy:", round(final_accuracy, 4))
print("Precision:", round(final_precision, 4))
print("Recall:", round(final_recall, 4))
print("F1 Score:", round(final_f1, 4))
```

```python
import joblib

model_package = {
    "model": final_model,
    "model_name": "Naive Bayes",
    "label_encoder": label_encoder,
    "features": list(X.columns),
    "accuracy": final_accuracy,
    "precision": final_precision,
    "recall": final_recall,
    "f1_score": final_f1,
    "medical_disclaimer": "This app is an educational AI prototype and does not provide medical diagnosis or prescription advice."
}

joblib.dump(model_package, "disease_prediction_model.pkl")

print("Final model saved successfully!")
print("Saved file: disease_prediction_model.pkl")
```
